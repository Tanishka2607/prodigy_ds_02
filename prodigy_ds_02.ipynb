```python
data['Age'] = data['Age'].fillna(data['Age'].median())
data
```

```python
data1['Age'] = data1['Age'].fillna(data1['Age'].median())
data1
```

```python
data['Embarked'] = data['Embarked'].fillna(method = 'ffill')
data
```

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data
```

```python
data1['Sex'] = data1['Sex'].map({'male': 0, 'female': 1})
data1
```

```python
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
data
```

```python
data1['Embarked'] = data1['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
data
```

```python
data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
data
```

```python
data1.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
data1
```
